```python
def change_columns_headers(df):
    '''enter dataframe as variable'''
    df.columns = df.columns.str.lower()
    return df
```

```python
def change_column_name(df, column_old_name, column_new_name):
    return df.rename(columns={column_old_name:column_new_name},inplace=True)
```

```python
def rearange_columns(df,list_of_new_column_names):
    '''as second argument put a list of new column names'''
    df = df[list_of_new_column_names]
    return df
```
